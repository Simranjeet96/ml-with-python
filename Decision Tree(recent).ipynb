```python
from sklearn import datasets
from sklearn import tree
import pydotplus
```

```python
iris = datasets.load_iris()
clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score as cv
scores = cv(clf , iris.data , iris.target , cv = 5)
scores.mean()
```

```python
clf.fit(iris.data ,  iris.target)
```

```python
dot_data=tree.export_graphviz(clf,feature_names=iris.feature_names,out_file=None,class_names=iris.target_names,filled=True,rounded=True,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data)
graph.write_jpeg("iriisss")
```
