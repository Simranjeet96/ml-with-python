```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append([list(movie_reviews.words(fileid)),category])
documents
```

```python
a=["A","a"]
for i in a:
    i=i.lower()
a
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
for i in documents[1:10]:
    for j in i[0]:
        if(j in stop):
            i[0].remove(j)
        else:
            if len(j)==1:
                i[0].remove(j)
```

```python
freq=[]
flag=0
for row in documents[1:10]:
    for i in row[0]:
        if(len(i)!=1):
            flag=0
            for j in freq:
                if j[0]==i:
                    j[1]=j[1]+1
                    flag=1
            if(flag!=1):
                freq.append([i,1])
                
```
