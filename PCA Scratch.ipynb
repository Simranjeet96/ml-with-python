```python
import numpy as np
```

```python
mu1 = [0,0,0]
sd1 = [[1,0,0] , [0,1,0] , [0,0,1]]
sample1 = np.random.multivariate_normal(mu1,sd1,20)
print(sample1)
```

```python
mu2 = [1,1,1]
sd1 = [[1,0,0] , [0,1,0] , [0,0,1]]
sample2 = np.random.multivariate_normal(mu2,sd1,20)
print(sample1)
```

```python
%pylab inline

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111 , projection = '3d')
ax.plot(sample1[:,0] , sample1[:,1] , sample1[:,2] , 'x' , color = 'blue')
ax.plot(sample2[:,0] , sample2[:,1] , sample2[:,2] , 'o' , color = 'green')
plt.show()
```

```python
sample1 = sample1.T
sample2 = sample2.T

all_samples = np.concatenate((sample1,sample2) , axis = 1)
print(all_samples)
```

```python
cov_mat = np.cov(all_samples)
print(cov_mat)
```

```python
eig_values , eig_vec = np.linalg.eig(cov_mat)
print(eig_values)
print(eig_vec)
```

```python
mat_W = np.array([eig_vec[:,1] , eig_vec[:,2]])
print(mat_W)
```

```python
projected = mat_W.dot(all_samples)
projected.shape
```
