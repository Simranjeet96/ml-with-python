```python
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:,:2]
Y = iris.target
```

```python
svc = SVC(kernel = 'linear')
svc.fit(X,Y)
```

```python
#min and max
x1_min , x2_min = X[: , 0].min() - 1 , X[: , 1].min() - 1
x1_max , x2_max = X[: , 0].max() + 1 , X[: , 1].max() + 1

#calculaing step size
h1 = (x1_max - x1_min)/100
h2 = (x2_max - x2_min)/100

#range of points
x1range = np.arange(x1_min , x1_max , h1)
x2range = np.arange(x2_min , x2_max , h2)
x2range
```

```python
xx , yy = np.meshgrid(x1range , x2range)
yy,xx
```

```python
points = np.c_[xx.ravel() , yy.ravel()]
points
```

```python
pred = svc.predict(points)
print(pred.shape)

pred = pred.reshape(xx.shape)
print(pred.shape)
```

```python
plt.contourf(xx , yy , pred , cmap = plt.cm.copper ,alpha = 0.4)
plt.scatter(X[:,0] , X[:,1] , c = Y , cmap = plt.cm.Pastel2_r)
plt.show()
```

```python
svc = SVC(kernel = 'rbf',C = 1)
svc.fit(X,Y)
pred = svc.predict(points)
pred = pred.reshape(xx.shape)
plt.contourf(xx , yy , pred , cmap = plt.cm.copper ,alpha = 0.4)
plt.scatter(X[:,0] , X[:,1] , c = Y , cmap = plt.cm.Pastel2_r)
plt.show()
```

```python
svc = SVC(kernel = 'rbf',C = 1000000  , gamma = 0.5)
svc.fit(X,Y)
pred = svc.predict(points)
pred = pred.reshape(xx.shape)
plt.contourf(xx , yy , pred , cmap = plt.cm.copper ,alpha = 0.4)
plt.scatter(X[:,0] , X[:,1] , c = Y , cmap = plt.cm.Pastel2_r)
plt.show()
```
