```python
mu_vecc1 = np.array([1,1,1])
std_vec1 = np.array([[1,0,0], [0,1,0], [0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vecc1, std_vec1, 20)
```

```python
class2_sample
```

```python
%pylab inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
```

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot(class1_sample[:, 0], class1_sample[:,1], class1_sample[:,2], '*', color = 'blue')
ax.plot(class2_sample[:,0], class2_sample[:, 1], class2_sample[:,2], 'o', color = 'red')

plt.show()
```

```python
class1_sample = class1_sample.T
class1_sample
```

```python
class2_sample = class2_sample.T
all_samples = np.concatenate((class1_sample, class2_sample), axis = 1)
print(all_samples[:,:])
all_samples[:,:].shape
```

```python
cov_mat = np.cov([all_samples[0,:], all_samples[1,:], all_samples[2,:]])
cov_mat
```

```python
eig_val_cov, eig_vec_cov = np.linalg.eig(cov_mat)
eig_vec_cov
```

```python
eig_val_cov
```

```python
matrix_W = np.array([eig_vec_cov[:,0], eig_vec_cov[:,2]])
matrix_W
```

```python
projected = matrix_W.dot(all_samples)
projected.shape
```

```python
plt.plot(projected[0, 0:20], projected[1, 0:20], 'o', color = 'blue')
plt.plot(projected[0, 20:40], projected[1, 20:40], '*', color = 'red')

plt.show()
```
