```python

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
import matplotlib.pyplot as plt
plt.subplot(2, 2, 1)#it divides height and width and activates it

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 2, 4)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```
