```python
from nltk.corpus import stopwords
from nltk.stem import  PorterStemmer
ps = PorterStemmer()
stop = set(stopwords.words('english'))
document={}
for category in movie_reviews.categories():
    temp=[]
    for fileid in movie_reviews.fileids(category):
        for word in movie_reviews.words(fileid):
            if word.isalpha():
                a=ps.stem(word)
                if (a not in stop and len(a)!=1):
                    temp.append(a)
    document[category]=temp
```

```python
import nltk
document['pos']=nltk.FreqDist(document['pos'])
document['neg']=nltk.FreqDist(document['neg'])
```

```python
document['pos']=dict(document['pos'])
document['neg']=dict(document['neg'])
```
