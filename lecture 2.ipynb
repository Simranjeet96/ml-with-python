```python
import pandas as pd
titanic=pd.read_csv("titanic_train.csv")
titanic
```

```python
def getnumber(str):
    if(str=="male"):
        return 1
    else :
        return 2
titanic["Sex"]=titanic["Sex"].apply(getnumber)
del titanic["SibSp"]
del titanic["Cabin"]
del titanic["Parch"]
del titanic["Fare"]
del titanic["Ticket"]
del titanic["PassengerId"]
del titanic["Name"]
```
